### **1. What is Supervised Learning?**

**Supervised learning** is the most common branch of Artificial Intelligence. In simple terms, it is **learning by example**.

**In supervised learning,** we provide an algorithm with a dataset that contains both the "questions" and the "correct answers."

**The goal**: the model to learn the relationship between them so it can predict answers for new, unseen questions.

**How it works**

image.png


- **Labeled data:** The training dataset contains both **input features**, such as email text, and the correct labels, like **spam or not spam.**

- **Training:** The algorithm studies these labeled examples to learn the relationship between **inputs** and their correct **outputs**.

- **Prediction:** After training, the model can take new, unseen data and predict the most likely category or outcome.

**Key Characteristics**

- **Labeled data:** Every data point comes with a known outcome or correct label.

- **Goal:** Learn a mapping from input x to output y.

- **Feedback loop:** The model compares its predictions with the true labels and adjusts itself to reduce error over time.

#### **1.1 Anatomy of a Dataset: Features vs. Targets**

To build a model, we must distinguish between the data we use for prediction and the thing we want to predict.

- **Inputs (Features):** Data we use to make predictions (e.g., age, blood pressure)

- **Outputs (Targets):** What we want to predict (e.g., disease presence)

- **Goal:** Learn a function that maps inputs to outputs

**Mathematical Concept**

$$y = f(x) +$$

- $y$ - target variable
- $f(x)$ - features
- $$ - noise or error

**The Features ($X$)**

Features (also called independent variables or predictors) are the individual measurable properties of the data.
- **Notation**: Usually represented as a matrix $X$.
- **Example:** If predicting house prices, features would be square footage, number of bedrooms, and ZIP code.

**The Target ($y$)**

The target (also called the dependent variable or label) is the specific value or category you want to predict.
- **Notation:** Usually represented as a vector $y$.
- **Example:** The actual sale price of the house.
